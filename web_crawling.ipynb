```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd

# 크롬 옵션 설정
chrome_options = Options()
chrome_options.add_argument("--headless")  # 브라우저 창을 띄우지 않고 실행
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# 크롬 드라이버 경로 설정
driver_path = 'C:/Users/user/Downloads/chromedriver_win32/chromedriver.exe'
# 크롬드라이버 경로로 변경

# Service 객체 생성
service = Service(executable_path=driver_path)

# 드라이버 초기화
driver = webdriver.Chrome(service=service, options=chrome_options)

def get_search_results(query, num_pages=5):
    base_url = "https://search.naver.com/search.naver"
    all_restaurants = []

    for page in range(1, num_pages + 1):
        start = (page - 1) * 10 + 1  # 네이버는 페이지당 10개씩 표시
        params = {
            'where': 'local',  # 지역 검색
            'query': query,
            'start': start
        }
        # URL 생성
        url = f"{base_url}?where={params['where']}&query={params['query']}&start={params['start']}"
        print(f"Fetching page {page}: {url}")
        driver.get(url)
        time.sleep(2)  # 페이지 로딩 대기

        # 페이지 소스 가져오기
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        # 검색 결과 항목 찾기
        # 네이버의 HTML 구조에 따라 수정 필요
        local_list = soup.find('ul', class_='list_local')
        if not local_list:
            print("No local list found on this page.")
            continue

        for li in local_list.find_all('li', class_='bx'):
            # 식당 이름 추출
            name_tag = li.find('a', class_='link_title')
            name = name_tag.get_text(strip=True) if name_tag else 'No Name'

            # 주소 추출
            addr_tag = li.find('span', class_='address')
            address = addr_tag.get_text(strip=True) if addr_tag else 'No Address'

            all_restaurants.append({'name': name, 'address': address})

    return all_restaurants

def save_to_csv(restaurants, filename='gangdong_zero_pay_restaurants.csv'):
    df = pd.DataFrame(restaurants)
    df.to_csv(filename, index=False, encoding='utf-8-sig')
    print(f"Saved {len(restaurants)} records to {filename}")

if __name__ == "__main__":
    search_query = "강동구 제로페이 식당"  # 검색어 설정
    num_pages = 10  # 크롤링할 페이지 수 설정

    restaurants = get_search_results(search_query, num_pages=num_pages)
    print(f"Total restaurants found: {len(restaurants)}")
    save_to_csv(restaurants)

    # 드라이버 종료
    driver.quit()
```
