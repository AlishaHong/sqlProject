```python
import pandas as pd
import pymysql
from tqdm import tqdm
```

```python
from glob import glob

fileName = glob('*.csv')[1]
fileName
```

```python
df = pd.read_csv(fileName, header = 1)
df.head()
```

```python
# 데이터 전처리 
# 소계 행 삭제하기

df = df.drop(0, axis = 0)
```

```python
df
```

```python
# 컬럼명 변경 
print(df)
```

```python
# 컬럼명 변경

df.columns = ['지역','총 인구','한국인','외국인']
df.head()
```

```python
# 연결하기 

conn = pymysql.connect(host = '127.0.0.1', user='root', password='1030',db='soloDB', charset='utf8')
cur = conn.cursor()
```

```python
# 데이터베이스와 테이블 생성하기 

def create():
    cur.execute("drop database if exists seoup_pop ")
    cur.execute("create database seoup_pop")
    cur.execute("use seoup_pop")
    cur.execute("""CREATE TABLE population(
                    지역 VARCHAR(255),
                    총인구 INT,
                    한국인 INT,
                    외국인 INT)""")
    conn.commit()
```

```python
def insert(row):
    for i in tqdm(range(row)):
        values = df.iloc[i].tolist()
        indata = "INSERT INTO population VALUES('{}', {}, {}, {})".format(*values)
        cur.execute(indata)
    conn.commit()
```

```python
create()
```
