```python
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://root:0000@127.0.0.1/BankDB')

df = pd.read_sql("SELECT * FROM CustomerAccount", engine)

df.head()
```

```python
def table_df(table_name, engine):

    query = f"SELECT * FROM {table_name}"
    
    df = pd.read_sql(query, engine)
    
    return df
```

```python
df = table_df('TransactionHistory', engine)

df.head()
```
